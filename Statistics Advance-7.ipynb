Q1. Write a Python function that takes in two arrays of data and calculates the F-value for a variance ratio
test. The function should return the F-value and the corresponding p-value for the test.

```python
import numpy as np
from scipy.stats import f_oneway

def calculate_f_value(array1, array2):
    """
    Calculate the F-value and p-value for a variance ratio test (ANOVA).

    Parameters:
    - array1: First array of data
    - array2: Second array of data

    Returns:
    - f_value: F-value for the test
    - p_value: p-value for the test
    """

    # Perform one-way ANOVA test
    f_value, p_value = f_oneway(array1, array2)

    return f_value, p_value

# Example usage:
data1 = np.array([1, 2, 3, 4, 5])
data2 = np.array([2, 3, 4, 5, 6])

f_value, p_value = calculate_f_value(data1, data2)

print("F-value:", f_value)
print("p-value:", p_value)
```

Q2. Given a significance level of 0.05 and the degrees of freedom for the numerator and denominator of an
F-distribution, write a Python function that returns the critical F-value for a two-tailed test.

```python
Answer = from scipy.stats import f

def get_critical_f_value(alpha, df_num, df_denom):
    """
    Calculate the critical F-value for a two-tailed test.

    Parameters:
    - alpha: Significance level
    - df_num: Degrees of freedom for the numerator
    - df_denom: Degrees of freedom for the denominator

    Returns:
    - critical_f_value: Critical F-value for the test
    """

    # Calculate the critical F-value using the ppf (percent point function) method
    # Percent point function (ppf) gives the critical value for the given probability (1 - alpha/2 for two-tailed test)
    critical_f_value = f.ppf(1 - alpha / 2, df_num, df_denom)

    return critical_f_value

# Example usage:
alpha = 0.05
df_num = 3  # Degrees of freedom for the numerator
df_denom = 10  # Degrees of freedom for the denominator

critical_f_value = get_critical_f_value(alpha, df_num, df_denom)

print("Critical F-value:", critical_f_value)
```

Q3. Write a Python program that generates random samples from two normal distributions with known

variances and uses an F-test to determine if the variances are equal. The program should output the F-
value, degrees of freedom, and p-value for the test.

```python
import numpy as np
from scipy.stats import f

def perform_f_test(sample1, sample2):
    """
    Perform an F-test to compare variances of two samples.

    Parameters:
    - sample1: First sample
    - sample2: Second sample

    Returns:
    - f_value: F-value for the test
    - df_num: Degrees of freedom for the numerator
    - df_denom: Degrees of freedom for the denominator
    - p_value: p-value for the test
    """

    # Calculate variances
    var1 = np.var(sample1, ddof=1)  # ddof=1 for unbiased estimation
    var2 = np.var(sample2, ddof=1)

    # Degrees of freedom
    n1 = len(sample1)
    n2 = len(sample2)
    df_num = n1 - 1
    df_denom = n2 - 1

    # Perform F-test
    f_value = var1 / var2 if var1 >= var2 else var2 / var1
    p_value = 2 * min(f.cdf(f_value, df_num, df_denom), 1 - f.cdf(f_value, df_num, df_denom))

    return f_value, df_num, df_denom, p_value

# Example usage:
np.random.seed(42)  # For reproducibility

# Generate random samples from normal distributions
sample1 = np.random.normal(loc=0, scale=1, size=30)
sample2 = np.random.normal(loc=0, scale=1.5, size=30)

# Perform F-test
f_value, df_num, df_denom, p_value = perform_f_test(sample1, sample2)

# Display results
print("F-value:", f_value)
print("Degrees of freedom (numerator):", df_num)
print("Degrees of freedom (denominator):", df_denom)
print("P-value:", p_value)
```

Q4.The variances of two populations are known to be 10 and 15. A sample of 12 observations is taken from
each population. Conduct an F-test at the 5% significance level to determine if the variances are
significantly different.

```python
import numpy as np
from scipy.stats import f

def perform_f_test(variance1, variance2, sample_size1, sample_size2):
    """
    Perform an F-test to compare variances of two populations.

    Parameters:
    - variance1: Variance of the first population
    - variance2: Variance of the second population
    - sample_size1: Sample size of the first population
    - sample_size2: Sample size of the second population

    Returns:
    - f_value: F-value for the test
    - df_num: Degrees of freedom for the numerator
    - df_denom: Degrees of freedom for the denominator
    - p_value: p-value for the test
    """

    # Degrees of freedom
    df_num = sample_size1 - 1
    df_denom = sample_size2 - 1

    # Perform F-test
    f_value = variance1 / variance2 if variance1 >= variance2 else variance2 / variance1
    p_value = 2 * min(f.cdf(f_value, df_num, df_denom), 1 - f.cdf(f_value, df_num, df_denom))

    return f_value, df_num, df_denom, p_value

# Given data
variance1 = 10
variance2 = 15
sample_size1 = 12
sample_size2 = 12
alpha = 0.05

# Perform F-test
f_value, df_num, df_denom, p_value = perform_f_test(variance1, variance2, sample_size1, sample_size2)

# Display results
print("F-value:", f_value)
print("Degrees of freedom (numerator):", df_num)
print("Degrees of freedom (denominator):", df_denom)
print("P-value:", p_value)

# Compare p-value to significance level
if p_value < alpha:
    print("Reject the null hypothesis. Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis. No significant difference in variances.")
```

Q5. A manufacturer claims that the variance of the diameter of a certain product is 0.005. A sample of 25
products is taken, and the sample variance is found to be 0.006. Conduct an F-test at the 1% significance
level to determine if the claim is justified.

```python
import numpy as np
from scipy.stats import f

def perform_f_test(claimed_variance, sample_variance, sample_size, alpha):
    """
    Perform an F-test to compare a claimed population variance to a sample variance.

    Parameters:
    - claimed_variance: The claimed population variance
    - sample_variance: The sample variance
    - sample_size: Sample size
    - alpha: Significance level

    Returns:
    - f_value: F-value for the test
    - df_num: Degrees of freedom for the numerator
    - df_denom: Degrees of freedom for the denominator
    - p_value: p-value for the test
    """

    # Degrees of freedom
    df_num = sample_size - 1
    df_denom = sample_size - 1

    # Perform F-test
    f_value = sample_variance / claimed_variance
    p_value = 2 * min(f.cdf(f_value, df_num, df_denom), 1 - f.cdf(f_value, df_num, df_denom))

    return f_value, df_num, df_denom, p_value

# Given data
claimed_variance = 0.005
sample_variance = 0.006
sample_size = 25
alpha = 0.01

# Perform F-test
f_value, df_num, df_denom, p_value = perform_f_test(claimed_variance, sample_variance, sample_size, alpha)

# Display results
print("F-value:", f_value)
print("Degrees of freedom (numerator):", df_num)
print("Degrees of freedom (denominator):", df_denom)
print("P-value:", p_value)

# Compare p-value to significance level
if p_value <= alpha:
    print("Reject the null hypothesis. The claim is not justified.")
else:
    print("Fail to reject the null hypothesis. The claim is justified.")
```

Q6. Write a Python function that takes in the degrees of freedom for the numerator and denominator of an
F-distribution and calculates the mean and variance of the distribution. The function should return the
mean and variance as a tuple.

```python
def calculate_f_distribution_params(df_num, df_denom):
    """
    Calculate the mean and variance of an F-distribution.

    Parameters:
    - df_num: Degrees of freedom for the numerator
    - df_denom: Degrees of freedom for the denominator

    Returns:
    - mean: Mean of the F-distribution
    - variance: Variance of the F-distribution
    """

    # Calculate mean
    mean = df_denom / (df_denom - 2)

    # Calculate variance
    variance = (2 * df_denom**2 * (df_num + df_denom - 2)) / (df_num * (df_denom - 2)**2 * (df_denom - 4))

    return mean, variance

# Example usage:
df_num = 3
df_denom = 10

mean, variance = calculate_f_distribution_params(df_num, df_denom)

print("Mean of F-distribution:", mean)
print("Variance of F-distribution:", variance)
```

Q7. A random sample of 10 measurements is taken from a normal population with unknown variance. The
sample variance is found to be 25. Another random sample of 15 measurements is taken from another
normal population with unknown variance, and the sample variance is found to be 20. Conduct an F-test
at the 10% significance level to determine if the variances are significantly different.

```python
import numpy as np
from scipy.stats import f

def perform_f_test(sample_variance1, sample_variance2, sample_size1, sample_size2, alpha):
    """
    Perform an F-test to compare variances of two samples.

    Parameters:
    - sample_variance1: Sample variance of the first sample
    - sample_variance2: Sample variance of the second sample
    - sample_size1: Sample size of the first sample
    - sample_size2: Sample size of the second sample
    - alpha: Significance level

    Returns:
    - f_value: F-value for the test
    - df_num: Degrees of freedom for the numerator
    - df_denom: Degrees of freedom for the denominator
    - p_value: p-value for the test
    """

    # Identify the sample with the larger variance
    if sample_variance1 >= sample_variance2:
        larger_sample_variance = sample_variance1
        smaller_sample_variance = sample_variance2
        larger_sample_size = sample_size1
        smaller_sample_size = sample_size2
    else:
        larger_sample_variance = sample_variance2
        smaller_sample_variance = sample_variance1
        larger_sample_size = sample_size2
        smaller_sample_size = sample_size1

    # Degrees of freedom
    df_num = larger_sample_size - 1
    df_denom = smaller_sample_size - 1

    # Perform F-test
    f_value = larger_sample_variance / smaller_sample_variance
    p_value = 2 * min(f.cdf(f_value, df_num, df_denom), 1 - f.cdf(f_value, df_num, df_denom))

    return f_value, df_num, df_denom, p_value

# Given data
sample_variance1 = 25
sample_variance2 = 20
sample_size1 = 10
sample_size2 = 15
alpha = 0.10

# Perform F-test
f_value, df_num, df_denom, p_value = perform_f_test(sample_variance1, sample_variance2, sample_size1, sample_size2, alpha)

# Display results
print("F-value:", f_value)
print("Degrees of freedom (numerator):", df_num)
print("Degrees of freedom (denominator):", df_denom)
print("P-value:", p_value)

# Compare p-value to significance level
if p_value <= alpha:
    print("Reject the null hypothesis. Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis. No significant difference in variances.")
```

Q8. The following data represent the waiting times in minutes at two different restaurants on a Saturday
night: Restaurant A: 24, 25, 28, 23, 22, 20, 27; Restaurant B: 31, 33, 35, 30, 32, 36. Conduct an F-test at the 5%
significance level to determine if the variances are significantly different.

```python
import numpy as np
from scipy.stats import f

def perform_f_test(sample1, sample2, alpha):
    """
    Perform an F-test to compare variances of two samples.

    Parameters:
    - sample1: Waiting times at Restaurant A
    - sample2: Waiting times at Restaurant B
    - alpha: Significance level

    Returns:
    - f_value: F-value for the test
    - df_num: Degrees of freedom for the numerator
    - df_denom: Degrees of freedom for the denominator
    - p_value: p-value for the test
    """

    # Calculate sample variances
    var1 = np.var(sample1, ddof=1)  # ddof=1 for unbiased estimation
    var2 = np.var(sample2, ddof=1)

    # Identify the sample with the larger variance
    if var1 >= var2:
        larger_sample_variance = var1
        smaller_sample_variance = var2
        larger_sample_size = len(sample1)
        smaller_sample_size = len(sample2)
    else:
        larger_sample_variance = var2
        smaller_sample_variance = var1
        larger_sample_size = len(sample2)
        smaller_sample_size = len(sample1)

    # Degrees of freedom
    df_num = larger_sample_size - 1
    df_denom = smaller_sample_size - 1

    # Perform F-test
    f_value = larger_sample_variance / smaller_sample_variance
    p_value = 2 * min(f.cdf(f_value, df_num, df_denom), 1 - f.cdf(f_value, df_num, df_denom))

    return f_value, df_num, df_denom, p_value

# Given data
waiting_times_A = np.array([24, 25, 28, 23, 22, 20, 27])
waiting_times_B = np.array([31, 33, 35, 30, 32, 36])
alpha = 0.05

# Perform F-test
f_value, df_num, df_denom, p_value = perform_f_test(waiting_times_A, waiting_times_B, alpha)

# Display results
print("F-value:", f_value)
print("Degrees of freedom (numerator):", df_num)
print("Degrees of freedom (denominator):", df_denom)
print("P-value:", p_value)

# Compare p-value to significance level
if p_value <= alpha:
    print("Reject the null hypothesis. Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis. No significant difference in variances.")
```

Q9. The following data represent the test scores of two groups of students: Group A: 80, 85, 90, 92, 87, 83;
Group B: 75, 78, 82, 79, 81, 84. Conduct an F-test at the 1% significance level to determine if the variances
are significantly different.

```python
import numpy as np
from scipy.stats import f

def perform_f_test(scores_group_A, scores_group_B, alpha):
    """
    Perform an F-test to compare variances of two groups of test scores.

    Parameters:
    - scores_group_A: Test scores of Group A
    - scores_group_B: Test scores of Group B
    - alpha: Significance level

    Returns:
    - f_value: F-value for the test
    - df_num: Degrees of freedom for the numerator
    - df_denom: Degrees of freedom for the denominator
    - p_value: p-value for the test
    """

    # Calculate sample variances
    var_group_A = np.var(scores_group_A, ddof=1)  # ddof=1 for unbiased estimation
    var_group_B = np.var(scores_group_B, ddof=1)

    # Identify the group with the larger variance
    if var_group_A >= var_group_B:
        larger_sample_variance = var_group_A
        smaller_sample_variance = var_group_B
        larger_sample_size = len(scores_group_A)
        smaller_sample_size = len(scores_group_B)
    else:
        larger_sample_variance = var_group_B
        smaller_sample_variance = var_group_A
        larger_sample_size = len(scores_group_B)
        smaller_sample_size = len(scores_group_A)

    # Degrees of freedom
    df_num = larger_sample_size - 1
    df_denom = smaller_sample_size - 1

    # Perform F-test
    f_value = larger_sample_variance / smaller_sample_variance
    p_value = 2 * min(f.cdf(f_value, df_num, df_denom), 1 - f.cdf(f_value, df_num, df_denom))

    return f_value, df_num, df_denom, p_value

# Given data
scores_group_A = np.array([80, 85, 90, 92, 87, 83])
scores_group_B = np.array([75, 78, 82, 79, 81, 84])
alpha = 0.01

# Perform F-test
f_value, df_num, df_denom, p_value = perform_f_test(scores_group_A, scores_group_B, alpha)

# Display results
print("F-value:", f_value)
print("Degrees of freedom (numerator):", df_num)
print("Degrees of freedom (denominator):", df_denom)
print("P-value:", p_value)

# Compare p-value to significance level
if p_value <= alpha:
    print("Reject the null hypothesis. Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis. No significant difference in variances.")
```
